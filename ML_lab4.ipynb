```
import numpy as np
import matplotlib.pyplot as plt
```

```
f_x = lambda x: x * np.cos(2 * x) # Функція 
step=0.00000000001
f_x_derivative = lambda x: ((f_x(x+step)-f_x(x))/step) # Похідна функції

x = np.linspace(-5, 4,1000) # Масив з заданим числом елемнтів рівномірно розподілений заданим інтервалом

plt.plot(x, f_x(x)) # бодуємо графік
plt.grid(True) # добавляємо сітку в графік
plt.show() # показуємо графік
```

```

def plot_gradient(x, y, x_vis, y_vis): # функція для побудови двох графіків
    plt.subplot(1,2,2) # перший графік
    plt.scatter(x_vis, y_vis, c = "b") # точки на графіку синім кольором (точки означають як Градієнтний спуск зміщується, щоб знайти мінімум функції)
    plt.plot(x, f_x(x), c = "r") 
    plt.title("Gradient Descent") # підпис графіка
    plt.show()

    plt.subplot(1,2,1) # другий графік
    plt.scatter(x_vis, y_vis, c = "b") 
    plt.plot(x,f_x(x), c = "r")
    plt.xlim([-3.5,-3.0]) # xlib - задається інтервал який буде виведений на графіку
    plt.title("Zoomed in Figure")
    plt.show()
    
```

```

def gradient_iterations(x_start, iterations, learning_rate):
    
    x_grad = [x_start] # список в який ми будемо записувати значення (по Х), для подальшої побудови графіків  
    y_grad = [f_x(x_start)] # список в який ми будемо записувати значення (по У), для подальшої побудови графіків

    for i in range(iterations):

        # Gradient Descent: X1 = X0 - (learning rate * dy/dx) 

        x_start_derivative = - f_x_derivative(x_start) # Знаходимо похідну в точці - dy/dx
        x_start += (learning_rate * x_start_derivative) # Похідну множимо на learning_rate і додаємо до точки    
        
        x_grad.append(x_start)
        y_grad.append(f_x(x_start))


    print("Start point: ", x_grad[0])
    print("Local minimum occurs at: {:.2f}".format(x_start))
    print("Number of steps: ",len(x_grad)-1)
    print()
    plot_gradient(x, f_x(x) ,x_grad, y_grad)

    return x_start
```

```
start_point =  -4.5 #@param {type:"number"}
num_of_iterations = 250 #@param {type:"slider", min:0, max:500, step:10}
learning_rate =  0.01 #@param {type:"number"}

gradient_iterations(start_point, num_of_iterations, learning_rate)
```

```
start_point =   3#@param {type:"number"}
num_of_iterations = 250 #@param {type:"slider", min:0, max:500, step:10}
learning_rate =  0.01 #@param {type:"number"}

gradient_iterations(start_point, num_of_iterations, learning_rate)
```

```
start_points = np.arange(-4.5, 3, 0.5)
global_min = []
for i in start_points:
    gradient_descent = gradient_iterations(i, num_of_iterations, learning_rate)
    global_min.append(gradient_descent)
```

```
global_minimum = min(global_min)
ind = global_min.index(global_minimum)
print("Start point: ", start_points[ind])
print("Global minimum occurs at: {:.2f}".format(global_minimum))
```

```
gradient_iterations(start_points[ind], num_of_iterations, learning_rate)
```
